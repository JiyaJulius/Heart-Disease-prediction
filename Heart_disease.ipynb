The input data consists of 13 features extracted from heart disease datasets, including attributes like age, sex, chest pain type (cp), resting blood pressure (trestbps), cholesterol level (chol), fasting blood sugar (fbs), resting electrocardiographic results (restecg), maximum heart rate achieved (thalach), exercise-induced angina (exang), ST depression induced by exercise (oldpeak), the slope of the peak exercise ST segment (slope), number of major vessels (ca), and thalassemia (thal).

```python
import pandas as pd

# Define file paths
txt_file_path = "processed.cleveland.txt"  # Text file path
csv_file_path = "processed.cleveland.csv"  # Output CSV file

# Define column names
column_names = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
                "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"]

# Read the text file with comma as a delimiter and assign column names
df1 = pd.read_csv(txt_file_path, delimiter=",", header=None, names=column_names)

# Save as CSV
df1.to_csv(csv_file_path, index=False)

print(f"File successfully converted to {csv_file_path}")
```

```python
df1.head()
```

```python
df1.replace("?", 0 , inplace=True)
df1
```

```python

# Count patients with and without heart disease
hd_positive = df1[df1["target"] > 0].shape[0]  # Assuming target > 0 means HD positive
hd_negative = df1[df1["target"] == 0].shape[0]  # Assuming target == 0 means HD negative

print(f"HD Positive: {hd_positive}")
print(f"HD Negative: {hd_negative}")
```

```python
#combined dataset
hd_positive = df_combined[df_combined["target"] > 0].shape[0]  # Assuming target > 0 means HD positive
hd_negative = df_combined[df_combined["target"] == 0].shape[0]  # Assuming target == 0 means HD negative

print(f"HD Positive: {hd_positive}")
print(f"HD Negative: {hd_negative}")
```

```python
# Convert 1, 2, 3, 4 to 1
df_combined['target'] = df_combined['target'].apply(lambda x: 1 if x > 0 else 0)

df_combined
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame for the bar plot data
data = pd.DataFrame({'Category': ['HD Positive', 'HD Negative'], 'Count': [hd_positive, hd_negative]})

# Use sns.barplot, providing the data source
# and the columns for x and y
sns.barplot(x='Category', y='Count', data=data)
plt.show()
```
